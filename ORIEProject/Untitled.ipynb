### Data exploration

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
weather = pd.read_csv('weather_hourly.csv')
energy = pd.read_csv('energy_hourly.csv')
```

#### Clean up dataframe, and filter out unusual energy demand

```python
weather['time'] = pd.to_datetime(weather['time'])

energy['time'] = pd.to_datetime(energy['slottime_GMT'])
energy = energy[['time','slotavg']]
energy = energy.rename(columns={'slotavg':'megawatts'})
mean, std =  np.mean(energy['megawatts']), np.std(energy['megawatts'])
energy = energy[(energy['megawatts']>(mean-3*std)) & (energy['megawatts']<(mean+3*std))]
```

#### Merging weather data (features) with energy demand (target)

```python
df = pd.merge(weather, energy, on='time', how='inner')
```

```python
X_data = df.drop(columns=['time','megawatts'])
y_data = df['megawatts']
```

#### Visualizing correlation matrix to check for multicollinearity

```python
corr_mat = X_data.corr()
corr_mat[corr_mat>0.7]
```

```python
drop_col = ['dewpoint_2m (°F)', 'apparent_temperature (°F)', 'pressure_msl (hPa)', 'precipitation (inch)',\
            'weathercode (wmo code)', 'cloudcover_low (%)', 'cloudcover_mid (%)', 'cloudcover_high (%)',\
            'shortwave_radiation (W/m²)', 'diffuse_radiation (W/m²)', 'direct_normal_irradiance (W/m²)',\
            'windspeed_100m (mp/h)']
```

#### New dataframe after filtering out highly correlated features

```python
X_data = X_data.drop(columns=drop_col)
X_data.head()
```

### Statistical analysis

#### Taking covariates and fitting an OLS

```python
model = sm.OLS(y_data, sm.add_constant(X_data)).fit()
model.summary()
```

```python
X_data_new = X_data.drop('snowfall (inch)', axis=1)
```

#### Doing the same thing with sliding windows to smooth out the outliers and missing data

```python
win_size = 24*7   # 7 day window
X_data_win = X_data.rolling(window=win_size).mean()[win_size-1:]
y_data_win = y_data.rolling(window=win_size).mean()[win_size-1:]
```

```python
win_model = sm.OLS(y_data_win, sm.add_constant(X_data_win)).fit()
win_model.summary()
```

```python
X_data_win = X_data_win.drop(columns=['relativehumidity_2m (%)', 'rain (inch)'])
```

#### Visualization of comparing energy demand between normal and sliding window 

```python
plt.plot(y_data, label='Normal')
plt.plot(y_data_win, label='Sliding window')
plt.legend()
plt.title('Normal vs. Sliding Window Energy Demand over Time')
plt.xlabel('# Hours since May 1, 2021')
plt.ylabel('Energy Demand (Mw)')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_validate
```

#### 80/20 train/test split

```python
X_train, X_test, y_train, y_test = train_test_split(X_data_new, y_data, test_size=0.2)
```

```python
model = LinearRegression().fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
line = range(21, 28)
plt.plot(line, line, c='red', label='y=x')
plt.xlabel('True demand (Mw)')
plt.ylabel('Predicted demand (Mw)')
plt.legend()
```

#### Accuracy of one train/test split after rounding to whole numbers

```python
sum(np.round(predictions) == np.round(y_test))/len(predictions)
```

#### 5-fold cross validation, not rounded so accuracy significiantly lower

```python
folds = 5  # 80/20 train/test split
output = cross_validate(model, X_data_new, y_data, return_train_score=True, cv=folds)

plt.plot(range(folds),output['train_score'], 'o-', label='train scores')
plt.plot(range(folds),output['test_score'], 'o-', label='test scores')
plt.hlines([output['test_score'].mean()], 0, folds-1, linestyle='dotted', color='C1')
plt.title(f'{folds}-fold validation')
plt.xticks(range(folds))
plt.grid(axis='y', linestyle='dotted')
plt.legend()
```

#### Average test scores of cv

```python
print(output['test_score'].mean())
```
