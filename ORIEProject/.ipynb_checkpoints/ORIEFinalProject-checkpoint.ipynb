```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Data ranges from 5/8/22 to 5/7/23

```python
weather = pd.read_csv('/content/ithaca 2022-05-08 to 2023-05-07.csv')
weather.head()
```

```python
energy = pd.read_csv('/content/Exported Data (3).csv')
energy.head(5)
```

```python
chilled_water = energy[energy['pointTitle'] == 'Chilled Water'].reset_index(drop='true')
chilled_water.head()
```

```python
weather['chilled_water'] = chilled_water['slotavg']
weather.head()
```

```python
covariates = ['tempmax', 'tempmin', 'temp', 'feelslikemax', 'feelslikemin', 'feelslike', 'humidity', 'uvindex']
x = weather[covariates]
y = weather['chilled_water']
model = sm.OLS(y, sm.add_constant(x)).fit()
model.summary()
```

```python
y_prime = model.predict(sm.add_constant(weather[covariates]))
plt.plot(y)
plt.plot(y_prime)
```
