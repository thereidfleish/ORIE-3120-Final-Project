```python
import pandas as pd
```

```python
df = pd.read_csv('/content/CookHouse.csv')
```

Data Set Up

Total Power

```python
Total_power = df[df['pointTitle'] == 'Cook House']
Total_power.groupby(['slottime_GMT']).mean()
Total_power = Total_power['slotavg'].reset_index(drop=True)
df1 = Total_power.head(268)
df2 = Total_power.tail(97)
Total_power = df1.append(df2, ignore_index=True)
```

Dining Power

```python
Dining_power = df[df['pointTitle'] == 'Electric for Dining']
Dining_power.groupby(['slottime_GMT']).mean()
Dining_power = Dining_power['slotavg'].reset_index(drop=True)
df1 = Dining_power.head(268)
df2 = Dining_power.tail(97)
Dining_power = df1.append(df2, ignore_index=True)
```

House Power

```python
House_power = Total_power - Dining_power
```

Graphs

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Total_power,'-', label = 'Total Power')
plt.plot(House_power,'-', label = 'House Power')
plt.plot(Dining_power,'-', label = 'Dining Power')
plt.legend(loc="upper right")
plt.xlabel('Month')
plt.ylabel('Power (kW)')
plt.title('Power Usage vs. Month')
plt.xlim(0, 365)
plt.ylim(0, 140)
x_ticks = [31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]
x_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
plt.xticks(ticks=x_ticks, labels=x_labels)
```
